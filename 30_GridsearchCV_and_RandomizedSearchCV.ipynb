```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

We will be working on the breast cancer dataset

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label']=breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

```python
x=np.asarray(x)
y=np.asarray(y)
```

```python
print(x)
```

```python
print(y)
```

#**<h1>Gridsearch cv**

GridSearchCV is used for determining the best parameters for our model

```python
model=SVC()
```

```python
#hyperparameter dictionary
parameters={
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
#grid search
classifier=GridSearchCV(model,parameters,cv=5)  # 5 part banakr somthing like that
```

```python
#fittting data to our model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
# best parameter
best_parameter=classifier.best_params_
print(best_parameter)
```

```python
#highest accuracy
highest_accuracy=classifier.best_score_
print(highest_accuracy)
```

```python
#loading result into pandas datafeme
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result=result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10, 'kernel':'linear'}

#**RandomizedSearchCV**

```python
model=SVC()
```

```python
#hyperparameter dictionary
parameters={
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
classifier=RandomizedSearchCV(model,parameters,cv=5)
```

```python
#fittting data to our model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
# best parameter
best_parameter=classifier.best_params_
print(best_parameter)
```

```python
#highest accuracy
highest_accuracy=classifier.best_score_
print(highest_accuracy)
```

```python
#loading result into pandas datafeme
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result=result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

Highest Accuracy = 95.08%

Best Parameters = {'C':5, 'kernel':'linear'}
