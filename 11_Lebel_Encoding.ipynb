Label encoding: converting the labels into numeric form(means o english me hote hai unhe no me convert karte hai 0 or 1 etc)

```python
#importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Data set

```python
#loading csv data fron csv file in pandas dataframe
cancer_data=pd.read_csv('/content/drive/MyDrive/AllDataset/11 breast_cancer_data.csv')
```

```python
#first five row
cancer_data.head()
```

```python
#finding the counts of diffrent labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder fun
label_encod=LabelEncoder()
```

```python
labels=label_encod.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to the dataFrame
cancer_data['target']=labels
```

```python
cancer_data.head()
```

note : ab hamne diagonis column(label) ko numeric me convert kar diya in the form of target so now we can remove diagonisis col

0--->Benign    

1--->Malignant

upper dataset me diagonisis me  M  and  B present the alphabatically B phele atahai isliye ise 0 mila and M baad me isliye isko 1 mila hai

```python
cancer_data['target'].value_counts()
```
